```python
# master_dataframe.isnull(): This part of the code creates a Boolean DataFrame of the same shape as master_dataframe, where each
# element is True if the corresponding element in the original DataFrame is NaN (missing) and False otherwise.

# .sum(): This part of the code sums the True values (which represent missing values) in the Boolean DataFrame for each column. 
# As a result, you get a Series with the count of missing values for each column.

master_dataframe.isnull().sum()
```

```python
# / len(master_dataframe): This part of the code divides the count of missing values for each column by the total number of rows
# in the DataFrame (len(master_dataframe)) to calculate the fraction of missing values as a percentage.

# * 100: Finally, this part of the code multiplies the fraction by 100 to convert it into a percentage.

# In summary, for each column in master_dataframe, the code counts the number of missing values (NaNs), divides it by the total
# number of rows, and then multiplies by 100 to express it as a percentage. This way, we get the percentage of missing values 
# for each column in the DataFrame.

# Calculate the percentage of missing values for each column
missing_percentage = (master_dataframe.isnull().sum() / len(master_dataframe)) * 100

# missing_percentage is a Pandas Series where the index represents column names, and the values represent the percentage of 
# missing values for each column.
missing_percentage
```

12. Create the ‘main_category’ column by extracting the main sector using the column ‘category_list’.
Hint - Use the Lambda function or string function.

```python
# To create a new 'main_category' column by extracting the main sector from the 'category_list' column in the master_dataframe, 
# we use the apply method with a lambda function. We need to define a function to extract the main sector based on 
# our specific criteria. Here the 'category_list' column contains a delimited list of categories and we assume that
# the main sector is the first category in the list.
```

```python
# Define a function to extract the main sector
def extract_main_sector(category_list):
    # Split the category list by the delimiter (e.g., '|') and return the first element as the main sector
    categories = str(category_list).split('|')
    if len(categories) > 0:
        return categories[0]
    else:
        return None  # Return None if there are no categories

# Apply the function to create the 'main_category' column
master_dataframe['main_category'] = master_dataframe['category_list'].apply(extract_main_sector)

# Display the DataFrame with the new 'main_category' column
print(master_dataframe[['category_list', 'main_category']])
```

13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical
column as follow
a) If the amount is less than 5 Millions then recode as TypeA.
b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
c) If the amount is greater than 7 Millions then recode as TypeC.

```python
import pandas as pd

# Assuming you have a DataFrame 'master_dataframe' with 'raised_amount_usd' column
# Replace this with your actual DataFrame

def recode_amount(amount):
    if amount < 5000000:
        return 'TypeA'
    elif 5000000 <= amount < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'

# Apply the custom function to create a new categorical column 'raised_amount_category'
master_dataframe['raised_amount_category'] = master_dataframe['raised_amount_usd'].apply(recode_amount)

# Display the updated DataFrame
display(master_dataframe[['raised_amount_usd', 'raised_amount_category']])
```
